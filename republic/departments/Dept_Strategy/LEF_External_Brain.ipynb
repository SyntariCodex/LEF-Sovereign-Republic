```python
# 1. Setup & Imports
import pandas as pd
import json
import numpy as np
from datetime import datetime

print("âœ… Brain Online.")
```

```python
# 2. Load LEF Memory (Data Ingestion)
try:
    df_trades = pd.read_csv('trade_history.csv')
    df_knowledge = pd.read_csv('knowledge_stream.csv')
    print(f"âœ… Loaded {len(df_trades)} Trades and {len(df_knowledge)} Knowledge Items.")
except FileNotFoundError:
    print("âš ï¸ Data not found. Please upload CSVs to the runtime.")
    # Mock Data for Demo
    df_knowledge = pd.DataFrame({'sentiment_score': [0.1, 0.5, 0.9, -0.2]})
```

```python
# 3. Deep Thought (Simulation of ML Model)
# In a real scenario, you would use PyTorch/TensorFlow here.

# Calculate Average Sentiment of recent knowledge
avg_sentiment = df_knowledge['sentiment_score'].mean()

# Determine Regime
if avg_sentiment > 0.5:
    regime = "BULL"
    risk = 0.8
    msg = "Knowledge Stream indicates high optimism. Recommendation: Aggressive Deployment."
elif avg_sentiment < -0.2:
    regime = "BEAR"
    risk = 0.2
    msg = "Knowledge Stream indicates fear. Recommendation: Capital Preservation."
else:
    regime = "NEUTRAL"
    risk = 0.5
    msg = "Market is sideways. Recommendation: Arbitrage & Yield Farming."

print(f"ðŸ”® Prediction: {regime} (Risk Level: {risk})")
```

```python
# 4. Synapse Firing (Generate Insight)

insight = {
    "type": "STRATEGIC_UPDATE",
    "source": "EXTERNAL_BRAIN_COLAB",
    "timestamp": datetime.now().isoformat(),
    "payload": {
        "regime_override": regime,
        "risk_level": risk,
        "message": msg,
        "model_version": "v1.0-alpha"
    }
}

filename = "strategic_update.json"
with open(filename, 'w') as f:
    json.dump(insight, f, indent=4)

print(f"âœ… Insight generated: {filename}")
print("â¬‡ï¸ Download this file and drop it into 'The_Bridge/Inbox'.")
```
